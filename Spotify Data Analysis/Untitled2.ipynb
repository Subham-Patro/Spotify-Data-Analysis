```python
df["release_date"] = pd.to_datetime(df["release_date"],errors='coerce')
df.set_index("release_date" , inplace = True)
```

```python
df
```

```python
df[['artists']].iloc[18]
```

```python
df['duration'] = df['duration_ms'].apply(lambda x : round(x/1000))
df.drop('duration_ms', inplace = True, axis=1)
```

```python
df.duration.head()
```

```python
df_corr = df.drop(["key","mode","explicit"],axis=1).corr(numeric_only=True)
plt.figure(figsize=(14,6))
heatmap = sns.heatmap(df_corr ,annot=True,fmt=".1g",vmin=-2,vmax=1,center =0,linecolor ="Black",cmap="inferno")
heatmap.set_title("Correlation Heatmap between variables")
```

```python
sample_df = df.sample(int (0.004*len(df)))
print(len(sample_df))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data =sample_df ,y = 'loudness' , x='energy' , color='c').set(title="loudness vs energy correlation")
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data =sample_df ,y = 'popularity' , x='acousticness' , color='c').set(title="popularity vs acousticness correlation")
```

```python
df['dates']=df.index.get_level_values('release_date')
df.dates=pd.to_datetime(df.dates)
years = df.dates.dt.year
```

```python
# sns.histplot(years,discrete= True, aspect = 2 , height=5,kind = 'hist').set(title="Number of songs per year")
plt.figure(figsize=(20,8))
sns.histplot(years, discrete=True, bins=len(set(years))).set(title="Number of songs per year")
```

```python
total_duration = df.duration
fig_dims=(18,7)
fig , ax = plt.subplots(figsize =fig_dims)
fig = sns.barplot(x=years,y=total_duration,ax=ax, errwidth=False).set(title="Years vs Duration")
plt.xticks(rotation=90)
```
